```python
movie_data = movie_data[['Genres','Price']]
movie_data
```

```python
movie_x = movie_data[['Genres']]
movie_y = movie_data[['Price']]
```

```python
xtrain,xtest,ytrain, ytest = train_test_split(movie_x,movie_y,test_size=0.20)
```

```python
xtrain.head()
```

```python
GenresEncoder = preprocessing.LabelEncoder()
GenresEncoder.fit(pd.DataFrame(xtrain[['Genres']]))
xtrain['Genres']=GenresEncoder.transform(xtrain[['Genres']])
```

```python
Genresscaler = StandardScaler()
Genresscaler.fit(pd.DataFrame(xtrain[['Genres']]))
xtrain['Genres']= Genresscaler.transform(xtrain[['Genres']])
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
GenresEncoder.fit(pd.DataFrame(xtest[['Genres']]))
xtest['Genres']=GenresEncoder.transform(xtest[['Genres']])
```

```python
Genresscaler.fit(pd.DataFrame(xtest[['Genres']]))
xtest['Genres']= Genresscaler.transform(xtest[['Genres']])
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(xtrain,ytrain)
```

```python
#Accuracy
print(accuracy_score(ytest,ypred))
```

**logistic regression is  possible to find the best accuracy**
